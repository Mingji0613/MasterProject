```python
import os
import numpy as np
import pandas as pd
# 导入库
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
# 展示中文
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False
# 创建文件夹，方便写入图片数据
root_path = '机器学习图片数据2'
if os.path.exists(root_path) != True: # path是文件夹或者文件的相对路径或者绝对路径
        os.mkdir(root_path)
```

## 数据预处理

```python
Genus = pd.read_excel("机器学习数据.xlsx",sheet_name='family', index_col=0)
Genus
```

```python
TYR = pd.read_excel("机器学习数据.xlsx",sheet_name='TYR')
TYR
```

```python
Genus = pd.DataFrame(Genus.values.T,columns=Genus.index,index=Genus.columns)
Genus
```

```python
Genus['Sample ID'] = Genus.index
Genus
```

```python
data =pd.merge(Genus,TYR)
data = data.drop(['Sample ID'], axis = 1)
data
```

```python
# 数据描述
print(data.info())
```

```python
# 查看缺失值
data.isnull().sum()
```

```python
col = data.columns.tolist()
col
```

## 相关性分析

```python
# 通过加入一个参数kind='reg'，seaborn可以添加一条最佳拟合直线和95%的置信带。
for i in range(len(col)-1):
    sns.pairplot(data, x_vars=col[i], y_vars=col[-1], height=7, aspect=0.8,kind='reg')
    plt.savefig(root_path + "/pairplot_"+str(col[i])+'.jpg')
    plt.show()
```

```python
# 相关系数矩阵 r(相关系数) = x和y的协方差/(x的标准差*y的标准差) == cov（x,y）/σx*σy
# 相关系数0~0.3弱相关0.3~0.6中等程度相关0.6~1强相关
m = data.corr()
m.to_csv(root_path + '/m.csv')
```

```python
# 筛选正相关特征
m_result = m['tyr(PPM)'].sort_values()
m_result.to_csv(root_path + '/m_result.csv')
```

```python
m_index = m_result.index.tolist()
m_values = m_result.tolist()
```

```python
m_set = {}
for i in range(len(m_index)-1):
    m_set[m_index[i]] = m_values[i]
m_set
```

```python
# 分别获取强相关，中等相关，弱相关列名
s_1, s_2, s_3 = [], [], []
for key,value in m_set.items():
    if 0 < value < 0.1:
        s_1.append(key)
    elif 0.1 < value <= 0.2:
        s_2.append(key)
    elif 0.2 < value <= 1:
        s_3.append(key)
    else:
        print(key,value)
```

## 模型预测与评估

###  建立基准（全特征属性）

```python
#导入数据集拆分工具
from sklearn.model_selection import train_test_split
# 数据集划分
y = data['tyr(PPM)'].astype(int)  # 以市盈率为标签
X = data.drop(['tyr(PPM)'], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler

#将数据拆分为训练集和验证集
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
#特征预处理
sc = StandardScaler()
#标准化
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 定义模型（多元线性回归）
from sklearn import linear_model
# 创建多元线性回归模型
model = linear_model.LinearRegression()
```

```python
# 训练模型
model.fit(X_train, y_train)
a = model.intercept_  # 截距
b = model.coef_  # 回归系数
```

```python
print("最佳拟合线:截距=", a, ",回归系数=", b)
print('常数:'+str(a))
score = model.score(X_test, y_test)
print('模型测试得分：'+str(score))
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
# 预测图
plt.figure(figsize = (12,7))
x = [i for i in range(len(y_test.tolist()))]
plt.plot(x, y_test, 'b', label="test data")
plt.plot(x, Y_pred, 'r', label="pred data")
# plt.xticks(x)
plt.legend(loc=2)
plt.title("测试集预测")
plt.savefig(root_path + "/测试集预测_基准.jpg")
plt.show()
```

```python
# 残差图
# 计算残差
residuals = y_test - Y_pred

# 绘制残差图
plt.figure(figsize = (12,7))
plt.scatter(Y_pred, residuals)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, c='r')
plt.title("残差")
plt.savefig(root_path + "/残差_基准.jpg")
plt.show()
```
